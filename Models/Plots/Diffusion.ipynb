```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from matplotlib import rc

rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
```

```python
t = np.linspace(0.0, 1.0, 100)
s = np.cos(4 * np.pi * t) + 2

fig, ax = plt.subplots(figsize=(6, 4), tight_layout=True)
ax.plot(t, s)

ax.set_xlabel(r'\textit{time (s)}', fontsize=12)
ax.set_ylabel('\\textit{Velocity (\N{DEGREE SIGN}/sec)}', fontsize=16)
ax.set_title(r'\TeX\ is Number $\displaystyle\sum_{n=1}^\infty'
             r'\frac{-e^{i\pi}}{3^n}$!', fontsize=16, color='r')
plt.show()
```

```python
def generate_trajectory(s):
    x0 = 0
    steps = np.random.randint(low=0, high=1+1, size=s)*2 - 1
    steps[0] = x0
    return np.cumsum(steps)
    
```

```python
plt.plot(np.arange(10000)/1000, generate_trajectory(10000))
```
